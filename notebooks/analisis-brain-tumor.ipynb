```python
base_path = "..\Proy_Biociencias\Proyecto_Biociencias\Brain Cancer - MRI dataset\Brain_Cancer raw MRI data\Brain_Cancer"
```

```python
df = pd.read_csv(r"..\Proy_Biociencias\Proyecto_Biociencias\Brain Cancer - MRI dataset\dataset.csv")
df
```

```python
# Listar las carpetas
clases = os.listdir(base_path)
print(f"Clases encontradas: {clases}")
```

```python
conteo = {}
for clase in clases:
    carpeta_clase = os.path.join(base_path, clase)
    imagenes = os.listdir(carpeta_clase)
    conteo[clase] = len(imagenes)
    print(f"{clase}: {len(imagenes)} im치genes")
```

```python
# Visualizar im치genes de ejemplo por clase
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
for i, clase in enumerate(clases):
    img_path = os.path.join(base_path, clase, os.listdir(os.path.join(base_path, clase))[0])
    img = Image.open(img_path)
    axs[i].imshow(img, cmap='gray')
    axs[i].set_title(clase)
    axs[i].axis('off')
plt.show()
```
